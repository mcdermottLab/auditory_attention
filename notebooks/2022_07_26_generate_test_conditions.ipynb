```python
import itertools
import pickle
```

```python
model_names = [
    'AttnTrackingControl',
    'AttnCNN',
    'AttnCNNPosSlope',
    'AttnCNNConstrained'
]

snrs = [-18, -15, -10, -5, 0, 5, 10, "clean"]

condition_pairs = list(itertools.product(model_names, snrs))

condition_dict = dict(zip(range(len(condition_pairs)), condition_pairs))
```

```python
condition_pairs
```

```python
out_name = "../attn_cnn_snr_eval_conditions.pkl"
```

```python
with open(out_name, 'wb') as handle:
    pickle.dump(condition_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open(out_name, 'rb') as handle:
    data = pickle.load(handle)
```

```python
data
```

```python
model_names = [
    'AttnTrackingControl',
    'AttnCNNConstrained',
    'AttnCNN',
    'AttnCNNPosSlope',
    'AttnCNNOnlyNorm'
]

snrs = [-18, -15, -10, -5, 0, 5, 10, "clean"]

num_bg_talkers = range(1, 5)

condition_pairs = list(itertools.product(model_names, snrs, num_bg_talkers))

condition_dict = dict(zip(range(len(condition_pairs)), condition_pairs))
```
