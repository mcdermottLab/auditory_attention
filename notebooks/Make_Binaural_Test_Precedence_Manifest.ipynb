# Generate conditions for precedence effect distractor test.

Test in style of Freymen et al. (1999) experiments 

Target at 0 aziuth    
Distractors at azimuth 0 and 60 degrees to right 
All signals at 0 elevation     

Distractor played at both locations at same level, with onset delay (precedence effect)
- equal power at both distractor locations
- 4 ms dely to induce percept of direction as origin of leading signal

Target set against distractor at desired SNR 

Will also generate diotic listening conditions   
No spatialization, just use precedecne effect on model 

## This Notebook
Generate location x snr conditions to get model performance and thresholds

### Make Freyman 1999 conditions 

```python
## Make co-located condition (F-F in paper) 
target_azim = 0
elevation = 0 
# create dict of trial dicts
snrs = [-12, -8, -4, 0]
front_cond_dict = {ix: {'target_loc': (0, 0),
                        'distract_loc': [0, 0],
                        'snr': snr,
                        'lead_channel': None,
                        'target_lead_channel': None,
                        'diotic': False}
                        for ix, snr in
                        enumerate(snrs)} 

n_front_cond_dict = len(front_cond_dict)
print(n_front_cond_dict)


# Assert keys contiguous
assert np.all(np.array(list(front_cond_dict.keys())) == np.arange(len(front_cond_dict)))

## Get F-R, F-RF, and F-FR  conditions 
target_azim = 0
elevation = 0 
distractor_azim = -60
lead_channel = ['center', 'side', None]
# target_lead_channel = ['center', 'side', None]

# create dict of trial dicts
freyman_cond_dict = {(ix + n_front_cond_dict): {'target_loc': (target_azim, elevation),
                                         'distract_loc': [distractor_azim, elevation],
                                         'snr': snr,
                                         'lead_channel': dist_channel,
                                         'target_lead_channel': None,
                                         'diotic': False}
                        for ix, (dist_channel, snr) in
                        enumerate(itertools.product(*[lead_channel, snrs]))} 

n_freyman_cond_dict = len(freyman_cond_dict)
print(n_freyman_cond_dict)

# append 
all_cond_dict = {**front_cond_dict, **freyman_cond_dict}
n_all_conds = len(all_cond_dict)
print(n_all_conds)

target_azim = 0
elevation = 0 
distractor_azim = 60
lead_channel = ['center', 'side']
# target_lead_channel = ['center', 'side', None]

# Get FR-XX conditions
# create dict of trial dicts
sim_targ_freyman_cond_dict = {(ix + n_all_conds): {'target_loc': (target_azim, elevation),
                                         'distract_loc': [distractor_azim, elevation],
                                         'snr': snr,
                                         'lead_channel': dist_channel,
                                         'target_lead_channel': 'center',
                                         'diotic': False}
                        for ix, (dist_channel, snr) in
                        enumerate(itertools.product(*[lead_channel, snrs]))} 

n_freyman_cond_dict = len(sim_targ_freyman_cond_dict)
print(n_freyman_cond_dict)

# append 
all_cond_dict = {**all_cond_dict, **sim_targ_freyman_cond_dict}
print(len(all_cond_dict))

# Assert keys contiguous
assert np.all(np.array(list(all_cond_dict.keys())) == np.arange(len(all_cond_dict)))

```

```python
# all_conds
# all conditions 
with open('binaural_test_manifests/freymen_1999_test_conds.pkl', 'wb') as f:
    pickle.dump(all_cond_dict, f)
```

### Make dev conditions 

```python
# First add conditions for confusion matrix 
snrs = np.linspace(3,-15,num=7).astype('int')

# Make diotic conditions 
locs = ['center', 'left', 'right']
clean_cond_diotic = {0: {'target_loc': 'center',
                      'distract_loc': 'center',
                      'snr': 'clean',
                      'lead_channel': 'center',
                      'diotic': True}
                      
                      }
diotic_cond_dict = {(ix + 1) : {'target_loc': target_loc,
                      'distract_loc': dist_loc,
                      'snr': snr,
                      'lead_channel': dist_loc,
                      'diotic': True}
                       for ix, (target_loc, dist_loc, snr) in
                        enumerate(itertools.product(*[locs, locs, snrs]))} 

n_diotic_conds = len(diotic_cond_dict) + 1 # add one to count clean diotic
print(n_diotic_conds)

target_azim = 0
elevation = 0 
distractor_azims = [10, 30, 60, 90] 
lead_channel = ['center', 'side', None]
# create dict of trial dicts
freyman_cond_dict = {(ix + n_diotic_conds): {'target_loc': (target_azim, elevation),
                                         'distract_loc': [dist_az, elevation],
                                         'snr': snr,
                                         'lead_channel': channel,
                                         'diotic': False}
                        for ix, (dist_az, channel, snr) in
                        enumerate(itertools.product(*[distractor_azims, lead_channel, snrs]))} 

n_freyman_cond_dict = len(freyman_cond_dict)
print(n_freyman_cond_dict)

all_cond_dict = {**clean_cond_diotic, **diotic_cond_dict, **freyman_cond_dict}
print(len(all_cond_dict))

# Assert keys contiguous
assert np.all(np.array(list(all_cond_dict.keys())) == np.arange(len(all_cond_dict)))

```
