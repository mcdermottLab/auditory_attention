```python
from pathlib import Path
import numpy as np 
import librosa 
from joblib import Parallel, delayed
```

```python
cv_path = Path('/om2/data/public/mozilla-CommonVoice-9.0/cv-corpus-9.0-2022-04-27/en/clips')

mp3s = list(cv_path.glob("*.mp3"))

```

```python
len(mp3s)
```

```python
from tqdm import tqdm


rates = Parallel(n_jobs=40)(delayed(librosa.get_samplerate)(mp3_path) for mp3_path in tqdm(mp3s))



# rates = [librosa.get_samplerate(mp3_path) for mp3_path in tqdm(mp3s)]
```
