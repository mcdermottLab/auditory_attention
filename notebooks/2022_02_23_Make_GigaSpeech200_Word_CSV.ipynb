```python
import torchaudio 
import torchaudio.transforms as T
```

```python
from IPython.display import Audio
```

```python
metadata = torchaudio.info(df.iloc[0,:].wav_path)
print(metadata)
```

```python
wav, sr = torchaudio.load(wav)
```

```python
Audio(wav, rate = sr)
```

```python
resampler = T.Resample(sr, 16000, lowpass_filter_width=64,
                           rolloff=0.9475937167399596, 
                           resampling_method="kaiser_window",beta=14.769656459379492, dtype=wav.dtype)
```

```python
new_wav = resampler(wav)
```

```python
Audio(new_wav, rate=16000)
```
