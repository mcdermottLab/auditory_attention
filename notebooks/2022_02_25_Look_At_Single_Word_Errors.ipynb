### Look at Librispeech model performance on top 200 Gigaspeech words

This model was trained on the Librispeech clean 100hr set, and achieved an average word error of 18% on the librispeech clean test set.  

```python
path = '../result/single_word_decode_test_output.csv' # test and dev are same set here 
df = pd.read_csv(path, sep='\t')
```

Looking at correct single word guesses:

```python
good_guesses = len(df[df['hyp'] == df['truth']]) 
percent_right = 100 * good_guesses / len(df)
print(f"{good_guesses} exact guesses; {round(percent_right,2)}% correct")
df[df['hyp'] == df['truth']]
```

Example of bad guesses

```python
df[df['hyp'] != df['truth']].head(50)
```

### GigaSpeech Model performance on top EnglishCCC words


Some of the target words of the English Consistent Confusion Corpus words are out of the GigaSpeech vocabulary, so we will only look at words the model knows. 

```python
path = '../result/english_ccc_single_word_decode_test_output.csv'
ecccdf = pd.read_csv(path, sep='\t')
```

```python
len(ecccdf)
```

Looking at correct single word guesses:

```python
good_guesses = ecccdf[(ecccdf['hyp'] == ecccdf['truth']) &  (ecccdf['truth'] != '<unk>')]
percent_right = 100 * len(good_guesses) / len(ecccdf)
print(f"{len(good_guesses)} exact guesses; {round(percent_right,2)}% correct")
good_guesses
```

Examples of incorrect guesses

```python
ecccdf[(ecccdf['hyp'] != ecccdf['truth']) &  (ecccdf['truth'] != '<unk>')].head(50)
```

## GigaSpeech Top 10 Decode

```python
path = '../result/gigaspeech_single_word_top_N_dev_beam-10-0.csv'
top_10 = pd.read_csv(path, sep='\t')
```

```python
top_10
```

```python
good_guesses = top_10[(top_10['hyp'] == top_10['truth']) & (top_10['beam']<=0)]
percent_right = 100 * len(good_guesses) / len(top_10.truth.unique())
print(f"{len(good_guesses)} exact guesses; {round(percent_right,2)}% correct")
good_guesses
```

```python
top_match_guesses = top_10[(top_10['hyp'] == top_10['truth'])]
percent_right = 100 * len(top_match_guesses) / len(top_10.truth.unique())
print(f"{len(top_match_guesses)} matched guesses; {round(percent_right,2)}% correct")
top_match_guesses
```

```python
close_guesses = top_10[top_10.apply(lambda x: x.truth in x.hyp, axis=1)]
percent_right = 100 * len(close_guesses.truth.unique()) / len(top_10.truth.unique())
print(f"{len(close_guesses)} included guesses; {round(percent_right,2)}% correct")
close_guesses
```
