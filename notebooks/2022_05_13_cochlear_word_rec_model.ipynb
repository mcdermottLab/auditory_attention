```python
import yaml

from pytorch_lightning import Trainer
```

```python
import sys 
sys.path.append('../')
from src.coch_word_rec_lightning import CochWordRecModule
```

```python
path = '../config/coch_word_rec/word_rec_1e-4lr.yaml'
config = yaml.load(open(path, 'r'), Loader=yaml.FullLoader)
```

```python
# config['data']['loader']['batch_size']
```

```python
config['n_jobs'] = 10
config['data']['audio']['rep_kwargs']['rep_on_gpu'] = 'True'
config['data']['loader']['batch_size'] = 8
```

```python
model = CochWordRecModule(config)
```

```python
trainer = Trainer(
    default_root_dir='../test_log_dump/',
   # log_every_n_steps = 10,
    limit_train_batches=0.001,
    limit_val_batches=0.0,

    num_nodes=1,
    gpus=1,
    accelerator="gpu",
)
```

```python
trainer.fit(model)
```
