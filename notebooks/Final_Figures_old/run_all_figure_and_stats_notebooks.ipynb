# Define Function to Run Notebooks
Define a function that takes a notebook file path as input, executes its contents, and saves the executed notebook.

```python
def run_notebook(notebook_path, output_path=None):
    """
    Executes a Jupyter notebook and saves the executed notebook.

    Parameters:
        notebook_path (str): Path to the input notebook file.
        output_path (str, optional): Path to save the executed notebook. 
                                     If None, overwrites the input notebook.
    """
    # Read the notebook
    with open(notebook_path, 'r', encoding='utf-8') as f:
        notebook = nbformat.read(f, as_version=4)

    # Create an ExecutePreprocessor instance
    ep = ExecutePreprocessor(timeout=600, kernel_name='python3')

    # Execute the notebook
    ep.preprocess(notebook, {'metadata': {'path': os.path.dirname(notebook_path)}})

    # Determine the output path
    if output_path is None:
        output_path = notebook_path

    # Save the executed notebook
    with open(output_path, 'w', encoding='utf-8') as f:
        nbformat.write(notebook, f)
```

```python
import os
import nbformat
from nbconvert import PythonExporter

# Directory containing the notebooks
notebooks_dir = "notebooks/Final_Figures"
output_dir = notebooks_dir  # Output directory for the converted scripts

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# List of notebook file paths to convert
notebook_paths = [
    "interaction_test_for_experiment_6.ipynb",
    "plot_experiment_1_and_sup_fig_1.ipynb",
    "plot_experiment_2.ipynb",
    "plot_experiment_3.ipynb",
    "plot_experiment_4_and_sup_fig_7.ipynb",
    "plot_experiment_5.ipynb",
    "plot_experiment_6.ipynb",
    "plot_experiment_7.ipynb",
    "plot_figure_4a_and_supplementary_figures_2_and_3.ipynb",
    "plot_figure_5_and_sup_figure_5.ipynb",
    "plot_figure_6.ipynb",
    "plot_supplementary_figure_4.ipynb"
]

# Convert each notebook to a Python script
for notebook_name in notebook_paths:
    notebook_path = os.path.join(notebooks_dir, notebook_name)
    output_path = os.path.join(output_dir, notebook_name.replace(".ipynb", ".py"))

    # Read and parse the notebook
    with open(notebook_path, "r") as nb_file:
        notebook_content = nbformat.read(nb_file, as_version=4)

    # Convert notebook to Python script
    exporter = PythonExporter()
    script, _ = exporter.from_notebook_node(notebook_content)

    # Remove cell markers (e.g., `# In[ ]:`)
    cleaned_script = "\n".join(
        line for line in script.splitlines() if not line.strip().startswith("# In[")
    )

    # Save the converted script
    with open(output_path, "w") as script_file:
        script_file.write(cleaned_script)

print(f"Converted notebooks saved to {output_dir}")
```
