```python
import h5py
import numpy as np
import os
import pandas as pd
from pathlib import Path
import pickle
import scipy.stats as stats
import soxr
import re
import IPython.display as ipd
```

```python
# Paths to files 

# stim_dir = Path("/om2/user/msaddler/data_word_recognition/speech_in_synthetic_textures/")
stim_dir = Path("/om/user/imgriff/datasets/speech_in_synthetic_textures/")
h5_files = list(stim_dir.glob("*/stim.hdf5"))
h5_files
```

```python
# get digit from file name
name = h5_files[0].parent.stem
int(re.findall(r'\d+', name)[0])
```

```python
# get word dict for models 
map_word_to_int = pickle.load( open("/om2/user/imgriff/datasets/commonvoice_9/en/cv_800_word_label_to_int_dict.pkl", "rb" )) 

map_int_to_word = {v: k for k, v in map_word_to_int.items()}
```

```python
# Look at h5 

h5 = h5py.File(h5_files[-1], "r")

# List keys
print(list(h5.keys()))
print(f"N examples: {len(h5['index_foreground'])}")
# Play audio for one of the examples 
ix = 12000
sr = h5["sr"][ix]
print("Foreground:", h5["index_foreground"][ix])
print("Texture:", h5["index_texture"][ix])
print("Signal:", h5["signal"][ix].shape)
print("Durration" , h5["dur"][ix])
print("SNR", h5["snr"][ix])
print(f"Target word: {map_int_to_word[h5['label_word_int'][ix]]}")
print("Cue")
ipd.display(ipd.Audio(h5["cue_signal"][ix], rate=sr, normalize=False))
print('Signal')
ipd.display(ipd.Audio(h5["signal"][ix], rate=sr, normalize=False))
h5.close()
```
