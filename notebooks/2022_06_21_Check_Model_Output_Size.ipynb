```python
import yaml
import numpy as np 
```

```python
path = '/om4/group/mcdermott/user/imgriff/projects/modmel/config/train_modmel.yaml'
config = yaml.load(open(path, 'r'), Loader=yaml.FullLoader)
```

```python
model_params = config['model']
```

```python
model_params
```

```python
n_layers = len(model_params['cnn_dim'])


kernel = [[2, 34], [2, 14], [5, 5], [5, 5], [2, 6], [2, 5], [2, 6]]
padding = model_params['padding']
stride = model_params['stride']
pool_stride = model_params['pool_stride']
pool_size = model_params['pool_size']

pool_stride = [[2, 4], [1, 4], [1, 4], [1, 1], [1, 1], [1, 1], [1, 4]]
pool_size = [[8, 13], [8, 13], [1, 13], [1, 13], [1, 1], [1, 1], [1, 13]]



w = 8000*10
h = 50

for l in range(n_layers):

    h = int(np.floor((h + 2 * padding[l] - kernel[l][0] ) / stride[l]) + 1)
    w = int(np.floor((w + 2 * padding[l] - kernel[l][1] ) / stride[l]) + 1)
    # pooling layers
    pool_pad = [size//2 for size in pool_size[l]]
    h = int(np.floor((h + 2 * pool_pad[0] - pool_size[l][0]) / pool_stride[l][0]) + 1)
    w = int(np.floor((w + 2 * pool_pad[1] - pool_size[l][1]) / pool_stride[l][1]) + 1)

print(f"height = {h}, width = {w}")
```
