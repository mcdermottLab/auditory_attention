```python
import pickle 
import numpy as np 
import itertools
```

```python
# Make dict pairing harmonic, inharmonic, and whispered conditions with clean or noisy conditions

# Make list of all possible combinations of conditions
harmonic = ['harmonic', 'inharmonic', 'whispered']
noise = ['clean', 'noisy']
all_conditions = list(itertools.product(harmonic, noise))
print(all_conditions)

cond_dict = {ix:{harm_type:True, 'clean_targets': noise == "clean"} for ix,(harm_type, noise) in enumerate(all_conditions)}
cond_dict
```

```python
# all_conds
# all conditions 
with open('timit_popham_2018_test_conditions.pkl', 'wb') as f:
    pickle.dump(cond_dict, f)
```
