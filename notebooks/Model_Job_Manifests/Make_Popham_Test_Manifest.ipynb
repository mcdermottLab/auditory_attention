```python
import pickle 
import numpy as np 
import itertools
from pathlib import Path
import pprint
```

# Current manifest generation 

### This makes manifest for expaneded test set. Manifest for exact human stim made in notebook:
* `gen_swc_popham_stim_for_expmnt_2024.ipynb` 

```python
# outline conditions 
target_harmonicity = ['harmonic', 'inharmonic', 'whispered']
distractor_harmonicity = ['harmonic', 'inharmonic', 'whispered', None]

conditions = list(itertools.product(target_harmonicity, distractor_harmonicity))

cond_dict = {ix:{"target_harmonicity":target_harm, 'distractor_harmonicity': distractor_harm} for ix, (target_harm, distractor_harm) in enumerate(conditions)}
pprint.pp(cond_dict)
```

```python
# all_conds
# all conditions 
with open('all_stim_swc_popham_exmpt_2024_cond_manifest.pkl', 'wb') as f:
    pickle.dump(cond_dict, f)
```
