```python
import pickle 
import numpy as np 
import itertools
from pathlib import Path
```

## Current manifest generation 

```python
## Get list of manifest
path_to_timit_dir = Path("/om2/user/imgriff/datasets/timit/")


manifests = ['/om2/user/imgriff/datasets/timit/harmonic_timit/all_targets_harmonic_single_distractor_0dB_SNR_jitter_fn_render.pdpkl' ,
'/om2/user/imgriff/datasets/timit/whispered_timit/all_targets_whispered_single_distractor_0dB_SNR.pdpkl',
'/om2/user/imgriff/datasets/timit/inharmonic_timit/all_targets_inharmonic_single_distractor_0dB_SNR.pdpkl']

noise = ['clean', 'noisy']
standard_conditions = list(itertools.product(manifests, noise))
print(standard_conditions)

std_cond_dict = {ix:{"manifest":manifest, 'clean_targets': noise == "clean"} for ix,(manifest, noise) in enumerate(standard_conditions)}
print(std_cond_dict)
n_std = len(std_cond_dict)

new_manifests = [path.as_posix() for path in path_to_timit_dir.glob('*/*0dB*.pdpkl') if 'all' not in path.as_posix()]
print(new_manifests)

new_cond_dict = {ix + n_std:{"manifest":manifest, 'clean_targets': False} for ix,manifest in enumerate(new_manifests)}

cond_dict = {**std_cond_dict, **new_cond_dict}
cond_dict
```

```python
# all_conds
# all conditions 
with open('timit_popham_2018_with_crossed_test_conditions.pkl', 'wb') as f:
    pickle.dump(cond_dict, f)
```

## Old way of generating manifests

```python
# Make dict pairing harmonic, inharmonic, and whispered conditions with clean or noisy conditions

# Make list of all possible combinations of conditions
harmonic = ['harmonic', 'inharmonic', 'whispered']
noise = ['clean', 'noisy']
all_conditions = list(itertools.product(harmonic, noise))
print(all_conditions)

cond_dict = {ix:{harm_type:True, 'clean_targets': noise == "clean"} for ix,(harm_type, noise) in enumerate(all_conditions)}
cond_dict
```
