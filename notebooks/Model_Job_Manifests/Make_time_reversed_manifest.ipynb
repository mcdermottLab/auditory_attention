```python
# Wanted SWC test stimuli conditions 

snrs = np.arange(-9,4,3)

cond_dict = {ix: {'snr':snr, 'time_reversed':True} for ix, snr in enumerate(snrs)}
cond_dict
```

```python
## Add non-reversed conditions for comparison 
n_conds = len(cond_dict)

new_conds = {(ix + n_conds): {'snr':snr, 'time_reversed':False} for ix, snr in enumerate(snrs)}


all_conds = {**cond_dict, **new_conds}
all_conds
```

```python
with open('binaural_test_manifests/time_reversed_test_manifest_1_distractor.pkl', 'wb') as f:
    pickle.dump(all_conds, f)
```
