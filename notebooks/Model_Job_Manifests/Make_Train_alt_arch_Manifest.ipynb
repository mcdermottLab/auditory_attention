# Make v10 alt arch configs

```python
wanted_configs = list(Path("config/arch_search/").glob("*v10*.yaml"))
len(wanted_configs)
```

```python
wanted_configs = [config for config in wanted_configs if "11" not in config.stem]
import re

natsort = lambda s: [int(t) if t.isdigit() else t.lower() for t in re.split(r'(\d+)', s.stem)]
wanted_configs = sorted(wanted_configs, key=natsort)
# ['a1', 'a2', 'a3', 'a10', 'a11', 'a22'] 


wanted_configs
```

```python
with open("binaural_train_manifests/v10_alt_arch_search_manifest.pkl", 'wb') as f:
    pickle.dump(wanted_configs, f, protocol=pickle.HIGHEST_PROTOCOL)
```
