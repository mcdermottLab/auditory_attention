# Load data


```python
### Load data

# includes summary for all participants and each model architecture (10 feature-gain models + 3 alternates) - need to average individual feature gain models
results_df = pd.read_csv('final_results_to_share/experiment_1_df.csv')

### Merge all feature-gain models into one group - will plot mean and SEM over architectures 
fba_models = results_df[(results_df.group == "Feature-gain Model") & (results_df.background_condition == '1-talker')].reset_index()


human_data = pd.read_pickle('final_results_dir/summary_2024_SWC_diotic_indiv_participant_results_195.pdpkl')
human_data['group'] = f'Human (N = {human_data.id_subject.nunique()})'

# rename human_df columns to match fba_models
human_data = human_data.rename(columns={'correct_mean': 'correct',
                                        'confusions_mean': 'confs',
                                        'adjusted_confusions_mean': 'confusions',
                                        'adjusted_accuracy_mean': 'accuracy',
                                         'condition': 'background_condition',
                                         })

human_data['model'] = human_data['id_subject']
results_for_ANOVA = pd.concat([fba_models, human_data], axis=0).reset_index(drop=True)
```

```python
from pingouin import rm_anova
import pingouin as pg

aov = pg.anova(dv='confusions', between='group', data=results_for_ANOVA,
               detailed=True)

aov.round(4)

# rm_anova_table = rm_anova(data=results_for_ANOVA, dv='confusions', subject='model', within=["group"], effsize='np2')
# print("ANOVA for model")
# rm_anova_table
```
