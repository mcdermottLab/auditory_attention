```python
import numpy as np
from torch.utils.data import Dataset, DataLoader

class RandomDataset(Dataset):
    def __getitem__(self, index):
        return np.random.randint(0, 1000, 3)

    def __len__(self):
        return 16
    
dataset = RandomDataset()
dataloader = DataLoader(dataset, batch_size=2, num_workers=4)
for batch in dataloader:
    print(batch)
```

```python
# This looks good too! Output is data ixs, np random sequence, python random sequence

import torch
from torch.utils.data import DataLoader, Dataset
import numpy as np
import random

class RandomDataset(Dataset):
    def __getitem__(self, ind):
        return ind, np.random.randint(1, 10000), random.randint(1, 10000)

    def __len__(self):
        return 24

dl = DataLoader(RandomDataset(), batch_size=4, shuffle=False, num_workers=6)

epochs = 2
for _ in range(epochs):
    for batch in dl:
        print(batch)
    print("====" * 10)
```
