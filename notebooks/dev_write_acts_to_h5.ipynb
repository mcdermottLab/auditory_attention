```python
import h5py 
import numpy as np
```

```python
with h5py.File("dummy_file2.h5", 'a') as f:
    for letter in "a":
        # check if letter is already in file
        if letter not in f.keys():
            f.create_dataset(f"{letter}/target", (10, 5), dtype="float32")
            f.create_dataset(f"{letter}/background", (10, 5), dtype="float32")

    for i in range(10):
        f["a"]['target'][i] = np.ones(5) * i 
        f["a"]['background'][i] = np.ones(5) * 3 * i
    
```

```python
h5 = h5py.File("dummy_file2.h5", 'r')
print(h5.keys())
print(h5['a']['target'][:])
print(h5['a']['background'][:])
h5.close()
```
