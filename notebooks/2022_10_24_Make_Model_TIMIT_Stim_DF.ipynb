```python
import pandas as pd
import pickle
import numpy as np 
import random
from tqdm import tqdm
from pathlib import Path
import re
```

```python
file_out_dir = list(Path('/om2/user/imgriff/datasets/timit/attn_task_dataframes/').glob('*dataset*.pdpkl'))
```

```python
list(file_out_dir)
```

```python
all_dfs = []

for file in file_out_dir:
    
    df = pd.read_pickle(file)
    df['dataset'] =  re.search('dataset_(\d+)', file.stem).group(0).split('_')[-1]
    all_dfs.append(df)
    
model_stim_df = pd.concat(all_dfs, ignore_index=True)
```

```python
model_stim_df.shape
```

```python
word_and_speaker_encodings = pickle.load( open( "/om2/user/imgriff/projects/End-to-end-ASR-Pytorch/word_and_speaker_encodings_jsinv3.pckl", "rb" )) 
class_map = word_and_speaker_encodings['word_idx_to_word']
word_to_ix_map = {val:key for key,val in class_map.items()}

```

```python
def get_ix_from_words(words):
    if isinstance(words, str):
        return words
    else:
        return [word_to_ix_map[word] for word in words]

model_stim_df['distractor_word_ints'] = model_stim_df['distractor_words'].apply(get_ix_from_words)
```

```python
out_path = Path('/om2/user/imgriff/datasets/timit/attn_task_dataframes/')

model_stim_df.to_pickle(out_path / 'timit_all_attn_stim_for_model_eval.pdpkl')
```
