```python
import sys
import IPython.display as ipd
import numpy as np
import pandas as pd
 
sys.path.append("/om2/user/msaddler/spatial_audio_pipeline")
from spatial_audio_util import util_audio as ua
 
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle("/om2/user/msaddler/spatial_audio_pipeline/assets/brir/mit_bldg46room1004/manifest_brir.pdpkl")
fn_pattern = '/om2/user/msaddler/spatial_audio_pipeline/assets/brir/mit_bldg46room1004/room{:04.0f}.hdf5'
```

```python
# index_room = np.random.randint(2000)
index_room = 0 
azim = df.src_azim.sample().iloc[0]
elev = df.src_elev.sample().iloc[0]
 
dfi = df[np.logical_and.reduce([
    df.index_room == index_room,
    df.src_elev == elev,
    df.src_azim.isin([azim, 360-azim]),
])]
index_brir = dfi.index_brir.values.tolist()
 
list_brir, sr = ua.load_brir(
    index_room=index_room,
    index_brir=index_brir,
    fn_pattern=fn_pattern,
    return_metadata=False,
    verbose=False,
)
sr = sr[0]
 
t = np.arange(0, 2, 1 / sr)
y = np.random.randn(*t.shape)
y = (np.sin(2 * np.pi * 2 * t) > 0) * y
 


for itr, brir in enumerate(list_brir):
    display_str = ""
    for k in ["index_room", "src_azim", "src_elev", "src_dist", "head_azim", "room_dim_xyz", "room_materials"]:
        display_str += f"{k}: {np.array(dfi.iloc[itr][k]).round().astype(int)} | "
    print(display_str[:-3])
    y_spatialized = ua.spatialize_sound(y, brir)
    ipd.display(ipd.Audio(rate=sr, data=y_spatialized.T, normalize=True))
```
