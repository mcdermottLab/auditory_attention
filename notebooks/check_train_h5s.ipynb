```python
import h5py
from tqdm.auto import tqdm
from pathlib import Path 
```

```python
train_paths = Path("/om/scratch/Mon/imgriff/datasets/spatial_audio_pipeline/assets/dataset_binaural_attn/v02/train").glob("*.hdf5*")

bad_files = []
errors = []
good_examples  = 0
for file in tqdm(train_paths):
    try:
        h5 = h5py.File(file, 'r')
        n_in_file = len(h5['target'])
        good_examples += len(h5['target'])

        h5.close()
    except Exception as e:
        errors.append(e)
        bad_files.append(file.stem)
```

```python
### Get paths with errors 

err_paths = Path("/om2/user/imgriff/datasets/spatial_audio_pipeline/generate_scenes/outLogs/").glob("*31949756*.err")

job_ixs = []
for fname in err_paths:
    with open(fname, "r") as f:
        # print(fname)
        if "error" in f.read().lower():
            job_ixs.append(int(fname.stem.split("_")[-1]))
    
```
