## Format conversion from Deepspeech2/Robust audio labels



```python
import json 
# from collections import Counter
```

```python
# Get librispeech labels 

words_labels = json.load(open('/om4/group/mcdermott/user/imgriff/projects/cocktail_party/deepspeech/librispeech_10K_labels.json'))
words = [word.upper() for word in words_labels[2:]] # Need to be upper case 

# save in E2E ASR format 

out_name = '../tests/sample_data/librispeech_words-10k.vocab'

with open(out_name, "w") as f:
            f.write("\n".join(words))

```

```python
# Check that they're the same lists 


with open(out_name, "r") as f:
    saved_words = f.read().splitlines()
```

```python
saved_words == words
```
