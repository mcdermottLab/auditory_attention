```python
# Download WordNet if not already downloaded
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

def remove_tenses(words):
    # Initialize a lemmatizer
    lemmatizer = nltk.stem.WordNetLemmatizer()

    # Define a function to get the part of speech for each word
    def get_wordnet_pos(word):
        tag = nltk.pos_tag([word])[0][1][0].upper()
        tag_dict = {"J": wordnet.ADJ, 
                    "N": wordnet.NOUN,
                    "V": wordnet.VERB, 
                    "R": wordnet.ADV}
        return tag_dict.get(tag, wordnet.NOUN)  # Default to noun if not found

    # Lemmatize each word based on its part of speech and store in a set to remove duplicates
    base_words = set()
    for word in words:
        base_word = lemmatizer.lemmatize(word, pos=get_wordnet_pos(word))
        base_words.add(base_word)

    return list(base_words)
```

```python
import string
```

```python
keep_dict = {}
rid_dict = {}
for word, sentences in full_dict.items():
    for i, sentence in enumerate(sentences):
        #strip punctution
        sentence = sentence.translate(str.maketrans('', '', string.punctuation))
        sentence = sentence.lower()
        sent_list = sentence.split(' ')
        try:
            word_loc = sent_list.index(word)
        except ValueError:
            word_loc = -1
        if 2 <= word_loc < len(sent_list) - 2:
            if word not in keep_dict:
                keep_dict[word] = [sentences[i],]
            else:
                keep_dict[word].append(sentences[i])
        else:
            if word not in rid_dict:
                rid_dict[word] = [sentences[i],]
            else:
                rid_dict[word].append(sentences[i])
```

```python
count = 0
b_count = 0
for word, sentences in keep_dict.items():
    count += len(sentences)
for word, sentences in rid_dict.items():
    b_count += len(sentences)
print('good sentences', count)
print('bad sentences', b_count)
```

```python
done_words = []
for word, sentences in keep_dict.items():
    sentences = set(sentences)
    if len(sentences) >= 2:
        done_words.append(word)
        sentences = list(sentences)[:2]
        keep_dict[word] = sentences
```

```python
final_rid_dict = rid_dict.copy()
for key in rid_dict.keys():
    if key in done_words:
        final_rid_dict.pop(key)
```
