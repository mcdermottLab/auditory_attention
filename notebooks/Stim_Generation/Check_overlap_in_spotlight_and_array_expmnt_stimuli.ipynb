# Find overlap in stimuli used for online spotlight and in person spatial release experiment 

```python
## Get spotlight stim 
sl_stim = pd.read_pickle("/om/user/imgriff/datasets/human_azim_spotlight_SWC_2024/human_azim_spotlight_stim_manifest.pdpkl")

## Get spatial release stim 
sr_path = "/om2/user/imgriff/datasets/spatial_audio_pipeline/assets/human_attn_experiment_v00/full_eval_trial_manifest_new_fnames.pdpkl"
sr_stim = pd.read_pickle(sr_path)
```

```python
import re

# Define a function to split the string
def split_string(s):
    return re.split(r'(\d{4}_)', s)[-1]

# Apply the function to the 'excerpt_src_fn' column in sl_stim dataframe
sl_stim['target_stim_basename'] = sl_stim['excerpt_src_fn'].apply(split_string)

sr_stim['target_stim_basename'] = sr_stim['src_fn'].apply(lambda x: Path(x).stem)

# get overlap 
print(len(set(sl_stim['target_stim_basename']).intersection(set(sr_stim['target_stim_basename']))))
print(len(set(sl_stim['target_stim_basename'])))
```
