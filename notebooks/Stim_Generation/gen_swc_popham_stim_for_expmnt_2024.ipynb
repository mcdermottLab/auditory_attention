### Make condition manifest for stimuli generation

````python
import itertools
````

````python
target_harmonicity = ['harmonic', 'inharmonic', 'whispered']
distractor_harmonicity = ['harmonic', 'inharmonic', 'whispered', None]

conditions = list(itertools.product(target_harmonicity, distractor_harmonicity))

cond_dict = {ix:{"condition":ix, "target_harmonicity":target_harm, 'distractor_harmonicity': distractor_harm} for ix, (target_harm, distractor_harm) in enumerate(conditions)}
pprint.pp(cond_dict)
````

````python
# all_conds
# all conditions 
with open('swc_popham_exmpt_2024_cond_manifest.pkl', 'wb') as f:
    pickle.dump(cond_dict, f)
````

### Break up conditions for reasonable stim generation time 

Generation takes roughly 1 minute/trial.   
Partition condition manifest to run batches of 30 trials per job. 

````python
total_trials_per_cond = 360 # 0 - 359 
conds_per_job = 30  
jobs_per_cond = total_trials_per_cond // conds_per_job 
print(f"{jobs_per_cond=}")

job_chain = list(itertools.chain.from_iterable(itertools.repeat(condition, jobs_per_cond) for condition in cond_dict.values()))
print(f"N total jobs = {len(job_chain)}")

job_dict = {ix:item for ix,item in enumerate(job_chain)}
````
