## Gen condition dicts for Hamrmonic, Inharmonic, and Whispered test 

```python
import importlib
import IPython.display as ipd

import numpy as np
import os
import pickle

import itertools
```

```python
harmonic_conds = ['harmonic', 'inharmonic', 'whispered']
clean_targets = [True, False]

all_conds = list(itertools.product(*[harmonic_conds, clean_targets]))

cond_dict = {ix: {harm:True, 'clean_targets':clean_targets} for ix, (harm, clean_targets)
            in enumerate(all_conds)}
cond_dict
```

```python
with open('../binaural_test_manifests/timit_harm_inharm_whisper_manifest.pkl', 'wb') as f:
    pickle.dump(cond_dict, f)
```
